```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movies.csv")
```

```python
df.head(20)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.fillna({
    "budget":df["budget"].mean()


})
```

```python
df =df.fillna({
"company": "Desconocido",
"released":"Desconocido",
"writer":"Desconocido",
"Country":"Desconocido",
        "rating" : "Desconocido",
                "genre":"Desconocido",
                "star":"Desconocido",
                "runtime":"Desconocido",
                "country":"Desconocido"


}
)

df
```

```python


df = df.dropna(subset=['gross'])
```

```python
df.isnull().sum()
```

```python
#analisis estadistica descriptiva

Promedio_gross = df["gross"].mean(),
Mediana_gross = df["gross"].median(),
Percentil75 = df["gross"].quantile([0.75])

print(Promedio_gross)
print(Mediana_gross)
print(Percentile75)



```

```python
data = {'Medida': ['Promedio_gross', 'Mediana_gross',"Percentile75"],
        'Valor': [78500541.01778312,20205757.0, 76016691.5 ]}
df_stats = pd.DataFrame(data)

```

```python
df_stats
```

```python
plt.style.use("ggplot")

ax = sns.barplot(data=df_stats, x="Medida", y="Valor", color="blue")

plt.title("Estadísticas descriptivas")

# Añadir etiquetas de datos dentro de las barras
for p in ax.patches:
    # Calcular la altura de la barra
    height = p.get_height()

    # Añadir el texto dentro de la barra
    ax.annotate(format(height, '.2f'),
                (p.get_x() + p.get_width() / 2., height / 2),
                ha='center', va='center',
                color='white',
plt.show()
```

```python
#Analisis univariado
plt.style.use("ggplot")
sns.histplot(data=df, x="year", color="green")
plt.title("Cantidad de Peliculas por año")
plt.show()
```

```python
sns.histplot(data=df, x="rating", color="green")
plt.title("Distribución por Rating")
plt.xticks(rotation=90)
plt.xlabel("Rating")
plt.ylabel("Frecuencia")
plt.show()
```

```python
sns.barplot(data=df, x="genre", y="score", color="Blue")
plt.title("Género por Calificación")
plt.xticks(rotation=90)
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Agrupación por director y recaudación total
top_directors = df.groupby('director')['gross'].sum().sort_values(ascending=False)

top_10_directors = top_directors.head(10)

fig = px.bar(x=top_10_directors.index, y=top_10_directors.values,
             labels={'x': 'Director', 'y': 'Recaudación Total'},
             title='Top 10 Directores con Mayor Recaudación Total')

fig.update_traces(marker_color='green')

fig.show()
```

```python
#Analisis multivariado
sns.pairplot(df)
plt.show()

```

```python
correlacion = df['gross'].corr(df['score'])
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="gross", y="score", color="green")
plt.title("Relación entre Ingresos brutos y Puntaje")
plt.text(0.5, 0.9, f"Correlación: {correlacion:.2f}",
         transform=plt.gca().transAxes, fontsize=12)
plt.show( )
```

```python
correlation_matrix = df[['votes', 'budget']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Matriz de Correlación entre Votes y Budget")
plt.show()
```
